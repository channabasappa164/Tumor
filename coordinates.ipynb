```python
import numpy as np
import cv2
from skimage import measure
import csv
import os
from google.colab import drive

drive.mount('/content/gdrive')

# Path to the image file
image_path = '/content/gdrive/MyDrive/yes/Y1.jpg'

# Load image from file
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if image is loaded successfully
if image is not None:
    # Perform image segmentation to extract the brain region
    segmented_image = image > 100  # Thresholding to segment the brain region (adjust threshold as needed)

    # Find contours of the segmented image
    contours = measure.find_contours(segmented_image, 0.5)

    # Calculate the radius of the segmented brain region
    props = measure.regionprops(measure.label(segmented_image))[0]
    radius = props.equivalent_diameter / 2  # Equivalent diameter gives the diameter, divide by 2 to get radius

    # Write predicted coordinates and radius to CSV
    num_coordinates_to_write = 50  # Change this to the desired number of coordinates
    with open('/content/gdrive/MyDrive/predicted_output.csv', 'w', newline='') as csvfile:
        fieldnames = ['x', 'y']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        # Sample coordinates based on the number of coordinates to write
        step = len(contours[0]) // num_coordinates_to_write
        for i in range(0, len(contours[0]), step):
            point = contours[0][i]
            # Write coordinates to CSV
            writer.writerow({'x': point[1], 'y': point[0]})

    print(f"Processed {os.path.basename(image_path)}: radius={radius}")
else:
    print(f"Error: Unable to load {image_path}")
```
