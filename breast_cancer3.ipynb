```python
#Stages - This code helps us to find stages of cancer.

import numpy as np
import tensorflow as tf
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler

# Load the breast cancer dataset
data = load_breast_cancer()

# Scale the data
scaler = StandardScaler()
X = scaler.fit_transform(data.data)
Y = data.target

# Build the model
N, D = X.shape
num_classes = len(data.target_names)
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(D,)),
    tf.keras.layers.Dense(num_classes, activation='softmax')
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
r = model.fit(X, Y, epochs=100)

# Function to predict cancer stage
def predict_cancer_stage():
    print("Enter tumor features:")
    features = []
    for feature_name in data.feature_names:
        feature_value = float(input(f"{feature_name}: "))
        features.append(feature_value)
    scaled_features = scaler.transform([features])
    prediction = model.predict(scaled_features)
    predicted_stage = np.argmax(prediction)
    print("Predicted cancer stage:", data.target_names[predicted_stage])

# Test the model on real-time data
while True:
    predict_cancer_stage()
    choice = input("Do you want to predict another tumor? (yes/no): ")
    if choice.lower() != 'yes':
        break
```

```python
#This code finds the type of tumor
import numpy as np
import tensorflow as tf
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the breast cancer dataset
data = load_breast_cancer()

# Scale the data
scaler = StandardScaler()
X = scaler.fit_transform(data.data)
Y = data.target

# Split data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Build the model
N, D = X_train.shape
num_classes = len(data.target_names)
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(D,)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(num_classes, activation='softmax')
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
r = model.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=100, batch_size=32)

# Function to predict cancer stage
def predict_cancer_stage():
    print("Enter tumor features:")
    features = []
    for feature_name in data.feature_names:
        feature_value = float(input(f"{feature_name}: "))
        features.append(feature_value)
    scaled_features = scaler.transform([features])
    prediction = model.predict(scaled_features)
    predicted_stage = np.argmax(prediction)
    print("Predicted cancer stage:", predicted_stage)

# Test the model on real-time data
while True:
    predict_cancer_stage()
    choice = input("Do you want to predict another tumor? (yes/no): ")
    if choice.lower() != 'yes':
        break
```

```python
import tensorflow as tf
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

try:
    %tensorflow_version 2.x
except Exception:
    pass

# Load the data
data = load_breast_cancer()

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(data.data, data.target, test_size=0.33)

# Scale the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Build the model
N, D = X_train.shape
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(D,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the model
r = model.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=100)

# Plot the training and validation loss
plt.plot(r.history['loss'], label='loss')
plt.plot(r.history['val_loss'], label='val_loss')
plt.legend()
plt.show()

# Evaluate the model on test set
test_loss, test_accuracy = model.evaluate(X_test, Y_test)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)

# Take new inputs and predict tumor type
while True:
    print("\nPlease provide the following tumor features:")
    features = []
    feature_names = data.feature_names
    for i, feature_name in enumerate(feature_names):
        value = float(input(f"{feature_name}: "))
        features.append(value)

    # Scale the new inputs
    scaled_features = scaler.transform([features])

    # Predict using the trained model
    prediction = model.predict(scaled_features)
    if prediction[0, 0] < 0.5:
        print("Predicted tumor type: Benign")
    else:
        print("Predicted tumor type: Malignant")

    choice = input("\nDo you want to predict another tumor type? (yes/no): ")
    if choice.lower() != 'yes':
        break
```
