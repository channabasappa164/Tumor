```python
#this code give x-axis,y-axis coordinates and radius of only one image


import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
import csv
import cv2
from google.colab import drive
from skimage import measure

drive.mount('/content/gdrive')

# Define the CNN model
def create_model(input_shape):
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(3))  # Output layer with 3 neurons for x, y coordinates, and radius
    return model

# Load image from file
image_path = '/content/gdrive/MyDrive/yes/Y1.jpg'  # Replace with the actual path to your image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Load image in grayscale

# Check if image is loaded successfully
if image is None:
    print("Error: Unable to load the image.")
else:
    # Perform image segmentation to extract the brain region
    segmented_image = image > 100  # Thresholding to segment the brain region (adjust threshold as needed)

    # Calculate the center of mass and radius of the segmented brain region
    props = measure.regionprops(measure.label(segmented_image))[0]
    center_of_mass = props.centroid
    radius = props.equivalent_diameter / 2  # Equivalent diameter gives the diameter, divide by 2 to get radius

    # Get image dimensions
    height, width = image.shape

    # Calculate coordinates relative to the left-bottom corner
    relative_x = center_of_mass[1]  # x-coordinate remains the same
    relative_y = height - center_of_mass[0]  # Invert y-coordinate

    # Write predicted coordinates and radius to CSV
    with open('/content/gdrive/MyDrive/predicted_coordinates.csv', 'w', newline='') as csvfile:
        fieldnames = ['x', 'y', 'radius']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        writer.writerow({'x': relative_x, 'y': relative_y, 'radius': radius})

    print("Predicted coordinates and radius written to predicted_coordinates.csv file.")
```

```python
#all the images x axis y axis and radius is foundout here
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
import csv
import cv2
from google.colab import drive
from skimage import measure
import os

drive.mount('/content/gdrive')

# Define the CNN model
def create_model(input_shape):
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(3))  # Output layer with 3 neurons for x, y coordinates, and radius
    return model

# Path to the folder containing images
folder_path = '/content/gdrive/MyDrive/yes'

# Get list of all files in the folder
image_files = os.listdir(folder_path)

# Open CSV file for writing
with open('/content/gdrive/MyDrive/predicted_coordinates_all.csv', 'w', newline='') as csvfile:
    fieldnames = ['Filename', 'x', 'y', 'radius']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    # Iterate through each image file
    for image_file in image_files:
        # Load image from file
        image_path = os.path.join(folder_path, image_file)
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

        # Check if image is loaded successfully
        if image is not None:
            # Perform image segmentation to extract the brain region
            segmented_image = image > 100  # Thresholding to segment the brain region (adjust threshold as needed)

            # Calculate the center of mass and radius of the segmented brain region
            props = measure.regionprops(measure.label(segmented_image))[0]
            center_of_mass = props.centroid
            radius = props.equivalent_diameter / 2  # Equivalent diameter gives the diameter, divide by 2 to get radius

            # Get image dimensions
            height, width = image.shape

            # Calculate coordinates relative to the left-bottom corner
            relative_x = center_of_mass[1]  # x-coordinate remains the same
            relative_y = height - center_of_mass[0]  # Invert y-coordinate

            # Write predicted coordinates and radius to CSV
            writer.writerow({'Filename': image_file, 'x': relative_x, 'y': relative_y, 'radius': radius})

            print(f"Processed {image_file}: x={relative_x}, y={relative_y}, radius={radius}")
        else:
            print(f"Error: Unable to load {image_file}")

print("All images processed. Predicted coordinates and radius written to predicted_coordinates_all.csv file.")
```
